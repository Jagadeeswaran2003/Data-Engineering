```python
filename = 'D:/ME204/Plum Project/Data/Clean/Plum_Project_Data.csv'
Plum_Project1 = pd.read_csv(filename)
```

```python
Plum_Project1
```

```python
conn = sqlite3.connect(os.path.join('../data', 'Plum_Projects.db'))
```

```python
Plum_Project1.to_sql('Plum_products' ,conn, if_exists='replace', index=False)
```

```python
Plum_product=conn.execute('''
    select * from Plum_products
''').fetchall()
```

```python
Plum_Product_DF = pd.DataFrame(Plum_product, columns=['Name_of_Products', 'Prices','Offer_Prices','Ratings', 'Reviews', 'Benefits','Skintypes','Positions', 'Offer_Products','Size_of_Products'])
```

```python
Plum_Product_DF
```

```python
(
    ggplot(data = Plum_Product_DF, mapping = aes(
                    x = 'Prices', 
                    y = 'Skintypes', 
                    color = 'Prices', 
                    size = 'Size_of_Products')
             ) +
        geom_point(alpha = 1) +
        scale_x_log10(),
)
```

```python
df = pd.DataFrame(Plum_Product_DF)
review_counts = df['Reviews'].value_counts()
plt.figure(figsize=(10,5))
plt.pie(review_counts, labels=review_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Product Reviews')
plt.show()
```

```python
(
    ggplot(
        data = Plum_Product_DF, 
        mapping = aes(
            x = 'Prices', 
            y = 'Offer_Prices', 
            color = 'Name_of_Products'
        )
    ) +
    geom_point() +
    geom_smooth(method = 'lm') +
    labs(
        title = 'Relationship Between Prices and Offer Prices',
        x = 'Original Price',
        y = 'Offer Price'
    ) +
    theme_minimal()
)
```

```python
(
    ggplot(data = Plum_Product_DF, 
       mapping = aes(x = 'Size_of_Products',
                     y = 'Reviews',
                     fill = 'Size_of_Products')
       ) +
        geom_violin(alpha=0.5) +
        geom_boxplot(alpha = 0.5) +
        guides(fill = 'none')  +
        labs(
            title = 'Distribution of Reviews by Product Size',
            y = 'Reviews',
            x = 'Size_of_Products'
        ) 
)
```

```python
ggplot (Plum_Product_DF, aes(x='Positions',y='Ratings')) + \
    geom_bar(star='identify',fill='Pink') + \
    ggtitle('Position by Ratings') + \
    xlab('Positions') + \
    ylab('Ratings') + \
    theme (axis_text_x=element_text(angle=45,hjust=1))
```

```python
Plum_Product_DF['Offer_Prices'] = Plum_Product_DF['Offer_Prices'].astype(str)  

plt.figure(figsize=(12, 10))

sns.violinplot(x="Size_of_Products", y="Offer_Prices", data=Plum_Product_DF, alpha=0.7)

plt.title('Size of Product Distribution of Offer Prices')
plt.xlabel('Offer_Prices')
plt.ylabel('Size_of_Products')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hexbin(Plum_Product_DF['Ratings'], Plum_Product_DF['Reviews'], gridsize=30, cmap='coolwarm', mincnt=1)
plt.colorbar(label="Count of Products")
plt.xlabel("Ratings")
plt.ylabel("Reviews")
plt.title("Ratings vs. Reviews")
plt.show()
```

```python
Plum_Product_DF = pd.DataFrame(Plum_Product_DF)

Offer_Products_counts = Plum_Product_DF['Offer_Products'].value_counts()

# Plot bar chart
plt.figure(figsize=(20,10))
Offer_Products_counts.sort_index().plot(kind='line', color='blue')
plt.title('Distribution of Product Ratings and Offer Product')
plt.xlabel('Offer_Products')
plt.ylabel('Ratings')
plt.xticks(rotation=0)  # Keeps the x-axis labels horizontal
plt.show()
```

```python
product_counts = Plum_Product_DF['Name_of_Products'].value_counts()
fig, ax = plt.subplots(figsize=(8, 6))
wedges, texts, autotexts = ax.pie(
    product_counts, 
    labels=product_counts.index, 
    autopct='%1.1f%%', 
    startangle=90,
    wedgeprops={'width': 0.4}
)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig.gca().add_artist(centre_circle)
plt.title('Distribution of Product Types', fontsize=16)
plt.tight_layout()
plt.show()
```
